```python
def add_company(inpt):
    if inpt == 'Samsung' or inpt == 'Mediacom' or inpt == 'Razer' or inpt == 'Microsoft' or inpt == 'Vero' or inpt == 'Xiaomi' or inpt == 'Chuwi' or inpt == 'Chuwi' or inpt == 'Fujitsu' or inpt == 'Google' or inpt == 'LG' or inpt == 'Huawei':
        return 'Other'
    else:
        return inpt

data['Company'] = data['Company'].apply(add_company)
    
```

```python
def set_processor(name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'Other'

data['cpu_name'] = data['cpu_name'].apply(set_processor)
```

```python
def set_os(inpt):
    inpt = str(inpt)
    if 'Windows' in inpt:
        return 'Windows'
    elif 'mac' in inpt.lower(): 
        return 'Mac'   
    elif inpt == 'Linux':
        return 'Linux'  
    else:
        return 'Other'


data = data.copy() 
data['OpSys'] = data['OpSys'].apply(set_os)


data['OpSys'].value_counts()
```

```python
data = data.drop(['laptop_ID', 'Inches', 'Cpu', 'ScreenResolution', 'Gpu','Product'], axis=1)
```

```python
data = pd.get_dummies(data)

data
```

```python
X = data.drop('Price_euros', axis = 1)
Y = data['Price_euros']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, Y_train)
    acc = model.score(X_test, Y_test)
    print(str(model)+'---->'+str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10, 50, 100],
              'criterion':['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit = grid_obj.fit(X_train, Y_train)

best_model = grid_fit.best_estimator_
```

```python
best_model
```

```python
best_model.score(X_test, Y_test)
```

```python
import pickle
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
